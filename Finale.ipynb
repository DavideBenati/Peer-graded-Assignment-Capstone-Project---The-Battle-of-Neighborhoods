```python
import numpy as np
import matplotlib.pyplot as plt

N = 23
ind = np.arange(N)  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)

yvals = PopulationPerDistrict.Schweizer
rects1 = ax.bar(ind, yvals, width, color='r')
zvals =  PopulationPerDistrict.Ausländer
rects2 = ax.bar(ind+width, zvals, width, color='g')


ax.set_ylabel('Population Distribution')
ax.set_xticks(ind+width)
ax.set_xticklabels( PopulationPerDistrict.District )
ax.legend( (rects1[0], rects2[0]), ('Swiss', 'Immigrants') )

def autolabel(rects):
    for rect in rects:
        h = rect.get_height()


autolabel(rects1)
autolabel(rects2)
plt.xticks(rotation=90)
plt.show()
```

```python
# instantiate a feature group for the districts in the dataframe
districts = folium.map.FeatureGroup()

for lat, lng, in zip(Basel_Sum_Data.Latitude, Basel_Sum_Data.Longitude):
    districts.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
Basel_Map.add_child(districts)
```

```python
from folium import plugins

# let's start again with a clean copy of the map of San Francisco
Basel_Map = folium.Map(location=[47.55, 7.6], zoom_start=13)

# instantiate a mark cluster object for the incidents in the dataframe
Basel_Pizzerias = plugins.MarkerCluster().add_to(Basel_Map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(Basel_Pizzeria.Latitude, Basel_Pizzeria.Longitude, Basel_Pizzeria.District):
    folium.Marker(
        location=[lat, lng],
        icon=None,
       popup=label,
    ).add_to(Basel_Pizzerias)

# display map
Basel_Map
```
